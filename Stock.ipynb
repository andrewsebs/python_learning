```python
import pandas as pd
import bar_chart_race as bcr
import ffmpeg
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
path="C://Users//harid//Downloads//StockData//OHLC.csv"
df=pd.read_csv(path)
df.head()
```

```python
df=df[['Symbol','Date','Close']]
df['Date']=pd.to_datetime(df['Date'].str.replace('.csv',''))
df.head()
```

```python
df=df.rename(columns={'Close':'LTP'})
df.head()
```

```python
df_filtered=df[df['LTP']>=300]
df_filtered.shape
```

```python
df_filtered=df_filtered.pivot_table(values='LTP', index='Date', columns='Symbol')
df_filtered.head()
```

```python
df_filtered=df_filtered.dropna(thresh=100,axis=1)
df_filtered.fillna(method='ffill', inplace=True)
df_filtered.fillna(method='bfill', inplace=True)
df_filtered.head()
```

```python
df_filtered=df_filtered.resample('M').last()
df_filtered.head()
```

```python
df_filtered.index = pd.to_datetime(df_filtered.index).strftime('%Y-%m')
df_filtered.head()
```

```python
bcr.bar_chart_race(df=df_filtered, title='10 most valuable script in NEPSE (Jan 2012 to Mar 2020)',
                   n_bars=10,filter_column_colors=True)
```
